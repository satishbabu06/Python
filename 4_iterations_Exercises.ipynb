```python
2.2 Write code to find only those numbers in a list that end with the digit 3. Create a new list with those numbers. 
    To accomplish this, you can convert the integers to strings and use indexing to access the last digit.
```

```python
x = [12, 43, 4, 1, 6, 343, 10, 34, 12, 93, 783, 330, 896, 1, 55]
y = []
for element in x:
    if str(element)[-1] == '3':
        y.append(element)
print(y)
```
