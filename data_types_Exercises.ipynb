```python
url = "https://raw.githubusercontent.com/data-bootcamp-v4/prework_data/main/students_performance.csv"
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/prework_data/main/students_performance.csv")
print(df.head())
data_types = df.dtypes
for column,dtype in data_types.items():
    if dtype == "object":
        column_type ="categorical"
    elif dtype in ["int64"]:
        if df[column].nunique() / len(df) < 0.05:
            column_type = "Categorical (Discrete)"
        else:
            column_type = "Numerical (Continuous)"
    print(f"Numerical column '{column}' is {column_type}")
        
```
