```python
def generation_calculator(birth_year):
    if birth_year < 1949:
        res = "Silent Generation !"
    elif birth_year > 1949 and birth_year < 1968 :
        res = "Baby Boomer"
    elif birth_year > 1969 and birth_year < 1980 :
        res = "Generation X"
    elif birth_year > 1981 and birth_year < 1993 :
        res = "Millennial"
    else:
        res = "Generation Z"
    return res        

dob = int(input("enter the birth year: "))
output = generation_calculator(dob)
print(f"entered birt year is {dob} and Generation is {output} enjoy Life." )
```
